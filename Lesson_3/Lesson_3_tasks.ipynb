#### 2. Задание
*Почему прямые не кажутся перпендикулярными?  
(Графики функций y = 3x + 1 и y = (-1/3)x + 1)*

```python
%matplotlib inline
%config InlineBackend.figure_format='svg'

import matplotlib.pyplot as plt
```

При настройках по умолчанию прямые действительно не кажутся перпендикулярными:

```python
x = np.linspace(-5, 5, 21)
y = 3*x+1
y2 = (-1/3)*x+1

plt.plot(x,y)
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

Дело в том, что здесь оси координат имеют разный масштаб,  
и сама плоскость построения графика - прямоугольная,  
т.е. вытянута по оси Х. Если мы приравняем масштабы осей  
и сделаем плоскость построения квадратной, то прямые  
становятся очевидно перпендикулярными:

```python
plt.figure(figsize=(4, 4))
plt.plot(x,y)
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y')
plt.axis(ymin=-5, ymax=5, xmin=-5, xmax=5)
plt.grid()
```

#### 3. Задание
*Напишите код на Python, реализующий построение графиков:  
окружности,  
эллипса,  
гиперболы.*

```python
# Параметры графиков:
a = 3
b = 1
x0 = -2
y0 = 1
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
fig.set_size_inches(12.6, 4)
plt.subplots_adjust(hspace=0.3)

# Окружность

x = np.linspace(-a + x0, a + x0, 200)
y1 = np.sqrt(a**2 - (x-x0)**2) + y0
y2 = -np.sqrt(a**2 - (x-x0)**2) + y0

ax1.plot(x, y1, color='blue')
ax1.plot(x, y2, color='blue')
ax1.axis([min(x)-2, max(x)+2, min(min(y1), min(y2))-2, max(max(y1), max(y2))+2])
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.grid()

# Эллипс

y1 = b/a * np.sqrt(a**2 - (x-x0)**2) + y0
y2 = -b/a * np.sqrt(a**2 - (x-x0)**2) + y0

ax2.plot(x, y1, color='blue')
ax2.plot(x, y2, color='blue')
ax2.axis([min(x)-2, max(x)+2, min(min(y1), min(y2))-2, max(max(y1), max(y2))+2])
ax2.set_xlabel('x')
ax2.grid()

# Гипербола

x1 = np.linspace(-3*a+x0, -a+x0, 200)
x2 = np.linspace(a+x0, 3*a+x0, 200)
y11 = b/a * np.sqrt((x1-x0)**2 - a**2) + y0
y12 = -b/a * np.sqrt((x1-x0)**2 - a**2) + y0
y21 = b/a * np.sqrt((x2-x0)**2 - a**2) + y0
y22 = -b/a * np.sqrt((x2-x0)**2 - a**2) + y0

ax3.plot(x1, y11, color='blue')
ax3.plot(x1, y12, color='blue')
ax3.plot(x2, y21, color='blue')
ax3.plot(x2, y22, color='blue')
ax3.axis([min(x1)-2, max(x2)+2, min(min(y12), min(y22))-2, max(max(y11), max(y21))+2])
ax3.set_xlabel('x')
ax3.grid()
```

3. Напишите код, который будет рисовать график отрезка прямой линии в полярных координатах.*

```python
def polar_to_cartesian(r, alpha):
    return (r*np.cos(alpha), r*np.sin(alpha))

def cartesian_to_polar(x, y):
    return (np.sqrt(x**2 + y**2), np.arctan(y/x))
```

```python
# Заданы полярные координаты:
r = 4
alpha = np.pi / 3
```

```python
# Декартовы координаты будут иметь вид:
x, y = polar_to_cartesian(r, alpha)
(x, y)
```

```python
# Перевод обратно в полярные координаты:
r, alpha = cartesian_to_polar(x, y)
(r, alpha)
```

#### 4. Задание
*Решите систему уравнений и неравенств:*  
1) exp(x) + x∙(1 – y) = 1  
y = x2 – 1
<br></br>
2) y = x2 – 1  
exp(x) + x∙(1 – y)  - 1 > 0

```python
from scipy.optimize import fsolve
```

Первая система уравнений:

```python
def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x * (1-y) - 1)
```

```python
# Построим графики функций для визуального определения количества корней
x = np.linspace(-5, 5, 200)
y1 = x**2 - 1
y2 = (np.exp(x) + x - 1) / x

plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

Имеем три пары корней в диапазоне х от -2 до 5

```python
x1, y1 = fsolve(equations, (-2,2))
x2, y2 = fsolve(equations, (2,5))
x3, y3 = fsolve(equations, (5,15))
```

```python
print(f'Первая пара корней: ({x1:.2f}, {y1:.2f})\n\
Вторая пара корней: ({x2:.2f}, {y2:.2f})\n\
Третья пара корней: ({x3:.2f}, {y3:.2f})')
```

Вторая система уравнений:

y = x2 – 1  
exp(x) + x∙(1 – y)  - 1 > 0

Неравенство системы повторяет уравнение из предыдущего примера.  
Разница только в том, что знак равенства заменён на знак ">".  
При выражении *y* из неравенства получим:  
y < (exp(x) + x - 1) / x  
Таким образом, решением системы будут интервалы синего графика  
на рисунке выше, которые лежат под оранжевым графиком.  
А именно, *х* будет находиться в диапазонах (-1.58, 2.62) и (4.2, *+∞*).  
Значения *y* получатся подставлением соответствующих значений *x* в  
уравнение *y* = *x*2 - 1
