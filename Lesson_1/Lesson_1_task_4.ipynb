##### Задание 4:
Постройте на одном графике две кривые y(x) для функции двух переменных y(k,x)=cos(k∙x),  
взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
%matplotlib inline
# %config InlineBackend.figure_format='svg'

import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-15, 15, 200)
k_1 = 1
k_2 = -0.5

plt.plot(x, np.cos(k_1 * x), label=f'y=cos({k_1}x)')
plt.plot(x, np.cos(k_2 * x), label=f'y=cos({k_2}x)')
plt.title('Графики функций y=cos(kx)')
plt.axis(ymin=-2, ymax=2)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()

plt.show()
```
